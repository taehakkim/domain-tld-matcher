```python
import pandas as pd

# Load the CSV file
result_df = pd.read_csv('D:\\도메인시트스플릿\\domainlist.csv', encoding='cp949')

result_df
```

```python
tld_list = [
    ".at", ".co.at", ".or.at", ".ch", ".li", ".ee", ".es",
    ".com.es", ".nom.es", ".org.es", ".gob.es", ".edu.es",
    ".fi", ".gr", "com.gr", "edu.gr", "net.gr", "org.gr",
    ".hr", "com.hr", ".ie", ".lu", ".lv", ".ro"
]
tld_list = [tld.lstrip('.') for tld in tld_list]
tld_list
```

```python
def has_special_ns(cell_value):
    domain_ns = set(str(cell_value).split(','))
    return not tld_list.isdisjoint(domain_ns)  # True if any match
```

```python
# Check if the value in the 3rd column (index 2) ends with any TLD in the list
matched_mask = result_df.iloc[:, 2].apply(
    lambda x: any(str(x).endswith(tld) for tld in tld_list)
)

# Filter rows
matched_rows = result_df[matched_mask]

# Output
matched_rows
```

```python
Expires = matched_rows['Expires'].drop_duplicates()
Expires
```

```python
combined = '; '.join(Expires)
print(combined)
```
